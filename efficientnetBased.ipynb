```python
def videoReader(path,n_frames=None):
    v_cap = cv2.VideoCapture(path)
    v_len = int(v_cap.get(cv2.CAP_PROP_FRAME_COUNT)) 
        # Pick 'n_frames' evenly spaced frames to sample
    if n_frames is None:
        sample = np.arange(0, v_len)
    else:
        sample = np.linspace(0, v_len - 1, n_frames).astype(int)
    filename = path.split('/')[-1]
    # Loop through frames
    #faces = []
    frames = []
    for j in range(v_len):
        success = v_cap.grab()
        if j in sample:
            # Load frame
            #cv2.VideoCapture.retrieve() : grab한 frame을 decode해서 반환
            success, frame = v_cap.retrieve()
            if not success:
                continue
            frame = cv2.resize(frame,(240,240))
            frames.append(frame)
    res_frame = frames[0]
    for j in range(1,len(frames)):
        res_frame = np.hstack((res_frame,frames[j]))
    filename = filename.split('.')[0]
    cv2.imwrite(SAVE_DIR + filename +'.jpg', res_frame)
```

```python
all_train_videos = glob.glob(os.path.join(TRAIN_DIR, '*.mp4'))
for path in tqdm(all_train_videos):
    videoReader(path,10)
```

```python
import json
all_train_images = glob.glob(os.path.join(SAVE_DIR, '*.jpg'))
# Get path of metadata.json
metadata_path = TRAIN_DIR + 'metadata.json'

# Get metadata
with open(metadata_path, 'r') as f:
    metadata = json.load(f)
```

```python
real = []
fake = []
for path in tqdm(all_train_images):
    file_name = path.split('/')[-1]
    file_name = file_name.split('.')[0] + '.mp4'

    if metadata[file_name]['label'] == 'FAKE':
        fake.append(path)
    else:
        real.append(path)
        

fake = random.sample(fake,len(real))
paths, Y =[], []

for x in real:
    paths.append(x)
    Y.append(0)
for x in fake:
    paths.append(x)
    Y.append(1)
```

```python
def shuffle(X,y):
    new_train=[]
    for m,n in zip(X,y):
        new_train.append([m,n])
    random.shuffle(new_train)
    X,y=[],[]
    for x in new_train:
        X.append(x[0])
        y.append(x[1])
    return X,y
```

```python
paths,y=shuffle(paths,Y)  
```

```python
def get_birghtness(img):
    return img/img.max()

def process_img(img):
    imgs=[]
    for x in range(10):
        imgs.append(get_birghtness(img[:,x*240:(x+1)*240,:]))
    return np.array(imgs)
```

```python
def gets(paths):
    al=[]
    for x in paths:
        al.append(process_img(cv2.cvtColor(cv2.imread(x),cv2.COLOR_BGR2RGB)))
    return al
def generator(paths,y,batch_size=16):
    while True:
        for x in range(len(paths)//batch_size):
            if x*batch_size+batch_size>len(paths):
                yield (np.array(gets(paths[x*batch_size:])),y[x*batch_size:])
            yield (np.array(gets(paths[x*batch_size:x*batch_size+batch_size])),
                   y[x*batch_size:x*batch_size+batch_size])
```
